```python
# 1. Write a Python Program to Add Two Matrices?
while True :
    r1 = int(input("Enter no of rowa in array 1 "))
    c1 = int(input("Enter no of coloums in array 1 "))
    r2 = int(input("Enter no of rowa in array 2 "))
    c2 = int(input("Enter no of coloums in array 2 "))
    l1 = []
    l2 = []
    l3 =[]
    if r1 != r2 or c1 != c2:
        print("Rows and coloums of both matrixes should be equal ")
    else:
        break
for i in range(r1):
    l1.append([])
    for j in range(c1):
        print("Enter",i ,j,"th Element of first Matrix ")
        l1[i].append(float(input("")))
for i in range(r2):
    l2.append([])
    for j in range(c2):
        print("Enter",i, j,"th Element of second Matrix ")
        l2[i].append(float(input("")))
for i in range(r1):
    l3.append([])
    for j in range(c1):
        l3[i].append(l1[i][j] +l2[i][j])
        print(l3[i][j],end=" ")
    print("\n")
```

```python
# 4. Write a Python Program to Sort Words in Alphabetic Order?
def swap(x,y):
    t = x
    x = y
    y = t
    return x , y

n = int(input("Enter no of elements "))
l = []
for i in range(n):
    l.append(input("Enter string "))
for i in range(len(l)):
    for j in range(i,len(l)):
        if ord(l[i][0]) > ord(l[j][0]):
            l[i] , l[j] = swap(l[i],l[j])
        elif ord(l[i][0]) == ord(l[j][0]):
            l1 = len(l[i])
            l2 = len(l[j])
            if l1 <= l2:
                for k in range(l1):
                    if ord(l[i][k]) > ord(l[j][k]):
                        l[i] , l[j] = swap(l[i],l[j])
                        break
            elif l1 > l2:
                for k in range(l2):
                    if ord(l[i][k]) > ord(l[j][k]):
                        l[i] , l[j] = swap(l[i],l[j])
                        break
print(l)
```

```python
# 5. Write a Python Program to Remove Punctuation From a String?
punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
s = input("Enter String ")
s = list(s)
r = ""
for i in s:
    if i not in punc:
        r = r + i
print(r)
```
